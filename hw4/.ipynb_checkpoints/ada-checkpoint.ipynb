```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import scale
import matplotlib
from matplotlib import pylab, gridspec, pyplot as plt
from sklearn.metrics import f1_score
import xgboost
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
train = pd.read_csv('train.csv', sep='\t', low_memory=False)
```

```python
train.dropna(inplace=True)
train.drop('DSSR', axis=1, inplace=True)
train.drop('pdb_chain', axis=1, inplace=True)
```

```python
train.head()
```

```python
y_train = np.ravel(train['mg'])
train.drop('mg', axis=1, inplace=True)
X_train = train[list(train.columns)].as_matrix()
```

```python
X_train
```

```python
y_train.shape, X_train.shape
```

```python
np.linspace(0.1, 1.0, 5)
```

```python
model = AdaBoostClassifier()
```

```python
grid = {'n_estimators': np.arange(200, 701, 100), 'learning_rate': np.linspace(0.1, 1.0, 5)}
gs = GridSearchCV(model, grid, cv=None, n_jobs=6, scoring='f1', verbose=2)
```

```python
gs.fit(X_train, y_train)
```

```python
gs.cv_results_ 
```

```python
gs.best_params_
```

```python
gs.best_score_
```

```python
X_tr, X_t, y_tr, y_t = train_test_split(X_train, y_train, test_size=0.33, random_state=42)
```

```python
n_estimators_grid = {'n_estimators': np.arange(700, 901, 50)}
n_estimators_grid
```

```python
gs_estimators = GridSearchCV(model, n_estimators_grid, cv=None, n_jobs=6, scoring='f1', verbose=2)
gs_estimators.fit(X_train, y_train)
```

```python
gs_estimators.best_params_, gs_estimators.best_score_
```

```python
test = pd.read_csv('test.csv', sep=',', low_memory=False)
test.head()
```

```python
test.dropna(inplace=True)
test.drop('Id', axis=1, inplace=True)
test.drop('index', axis=1, inplace=True)
test.drop('DSSR', axis=1, inplace=True)
test.drop('pdb_chain', axis=1, inplace=True)
```

```python
test.head()
```

```python
X_test = test.as_matrix()
X_test
```
